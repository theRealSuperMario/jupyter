```python
%load_ext nb_black
import numpy as np
import cv2
import skimage
from typing import *
```

```python
point_list = np.array([
    [0, 0],
    [1, 0],
    [0, 1],
    [1, 1]
])
```

```python
T = np.array([
    [1, 0, 0],
    [0, 0.6, 0],
    [0, 0, 1]
]).astype(np.float32)
point_list = np.array([
    [0, 0],
    [0, 1],
    [1, 1],
    [1, 0],
]).astype(np.float32).reshape((-1, 1, 2))

print(f"T : {T.shape}")
print(f"point_list : {point_list.shape}")

out_list = cv2.perspectiveTransform(point_list, T)

print(f"transformed points shape : {out_list.shape}")
```

```python
from matplotlib import pyplot as plt
```

```python
polygon1 = np.int32(out_list.reshape((-1, 2)) * 255 + 50)
polygons = [polygon1]

img = skimage.data.astronaut()
img_drawed = cv2.polylines(img, polygons, True, (0, 255, 255), 20)
plt.imshow(img_drawed)
```

# cv2.polylines

* Takes a `list` of `np.ndarray`s with shape `(N, 2)`.
* Each element in the list describes an individual polygon.
* The polygon points need to be integer image coordinates.
* To plot multiple polygons, just provide multiple point sets in the list. See example

### cv2.polylines - plotting single polygon

```python
polygon1 = np.array([
    [0, 0],
    [1, 0],
    [1, 1],
    [0, 1]
]).astype(np.float32).reshape((-1, 2))

polygon1 = np.int32(polygon1 * 255 + 50)

polygons = [polygon1]

img = skimage.data.astronaut()
img_drawed = cv2.polylines(img, polygons, True, (0, 255, 255), 20)
plt.imshow(img_drawed)
```

### cv2.polylines - plotting multiple polygons

```python
polygon1 = np.array([
    [0, 0],
    [1, 0],
    [1, 1],
    [0, 1]
]).astype(np.float32).reshape((-1, 2))
polygon2 = np.array([
    [0, 0],
    [1, 0],
    [1, 1],
    [0, 1]
]).astype(np.float32).reshape((-1, 2))

polygon1 = np.int32(polygon1 * 255 + 50)
polygon2 = np.int32(polygon2 * 125 + 350)

polygons = [polygon1, polygon2]

img = skimage.data.astronaut()
img_drawed = cv2.polylines(img, polygons, True, (0, 255, 255), 20)
plt.imshow(img_drawed)
```
