```python
import scipy
from scipy import signal
```

The system equation is given by

\begin{align}
    x_t = \beta x_{t - 1} + (1 - \beta) u_t
\end{align}
where $u_t$ is a general input function.

In the z-plane, the transfer function looks like this

\begin{align}
    H(z) = \frac{(1 - \beta)z^{1}}{-\beta z^{0} + z}
\end{align}

```python
beta = 0.9
b = [1 - beta, 0]
a = [1, -beta]

sys = signal.dlti(b, a)
```

```python
sys.to_tf()
```

```python
# Initial condition: x_0 = 0
tout, yout = sys.step(x0=0)
```

```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
plt.plot(tout, np.squeeze(yout[0]))
plt.xlabel(r"timestep $t$")
plt.ylabel(r"$x_t$")
plt.savefig("step_response_adam.png")
```
