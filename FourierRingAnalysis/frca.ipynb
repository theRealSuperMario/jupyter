## Gradually Increase Blur

```python
FRCs = []
Rs = []
sigmas = [1.5, 3, 6, 9, 12, 15]
for i, sigma in enumerate(sigmas):
    image1 = skimage.data.shepp_logan_phantom()
    image2 = image1.copy()
    image2 = skimage.filters.gaussian(image1, sigma=sigma)
    width_ring = 5
    polynom_degree = 20
    resol_square = False
    plot=True
    pathout = None
    frc, r = fourier_ring_correlation.analysis_frc_v3(image1, image2, width_ring, polynom_degree, resol_square, plot, pathout=pathout, prefix="pre", image_name="image_name")
    FRCs.append(frc)
    Rs.append(r)
```

```python
from supermariopy import plotting
from matplotlib import pyplot as plt
plotting.set_style()

colors = plt.cm.coolwarm(np.linspace(0, 1, len(sigmas)))[:, :3]

with plt.rc_context(plotting.NB_RC_PARAMS):
    for frc, r, s, c in zip(FRCs, Rs, sigmas, colors):
        plt.plot(r, frc, label=f"$\sigma ={s:1.02f}$", color=c)
    plt.legend(loc="upper right")
    plt.xlabel("$r$")
    plt.ylabel("FRC")
```
