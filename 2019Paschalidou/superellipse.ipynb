## Superellipse function

$x(\theta) = [
        a_1 \cos(\theta)^\epsilon, \\
        a_2 \sin(\theta)^\epsilon
]
$

# Adaptive Sampling - Model A

$
\theta_0 = 0 \\
\theta_i = \theta_{i-1} + \Delta_\theta(\theta_i) \\
\theta_N < \frac{\pi}{2} \\
\Delta_{\theta} (\theta) 
    = 
    \frac{D(\theta)}{\epsilon} 
    \sqrt{
        \frac{
            \cos(\theta)^2 \sin(\theta)^2
        }{
            a_1^2(\cos(\theta)^\epsilon)^2 \sin(\theta)^4 + a_2^2(\sin(\theta)^\epsilon)^2 \cos(\theta)^4
        }
    }
$
